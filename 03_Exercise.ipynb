```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from jcopml.pipeline import num_pipe, cat_pipe
from jcopml.utils import save_model, load_model
from jcopml.plot import plot_missing_value
from jcopml.feature_importance import mean_score_decrease
```

```python
df = pd.read_csv('/content/drive/MyDrive/00_History_Matching/Data/df_merged.csv')
df.head()
```

```python
import pandas as pd

# Initialize an empty list to store the split datasets
split_datasets = []

# Get unique values in the "case_num" column
unique_cases = df['case_num'].unique()

# Iterate over each unique case_num
for case in unique_cases:
    # Filter the data for the current case_num and the desired range of days
    case_data = df[(df['case_num'] == case) & (df['days'] >= 2500) & (df['days'] <= 3500)]
    
    # Append the split dataset to the list
    split_datasets.append(case_data)

# Concatenate the split datasets into a single DataFrame
Validation_df = pd.concat(split_datasets)

# Reset the index of the concatenated DataFrame
Validation_df.reset_index(drop=True, inplace=True)
```
